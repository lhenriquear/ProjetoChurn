```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
dx = pd.read_csv("Top Games.csv")
```

```python
dx.info()
```

```python
dx['description'] = dx['Name_URL'] + dx['Name'] + dx['Name1'] + dx['Score2']+ dx['Genre'] + dx['ReleaseDates1'] + dx['GameModes'] + dx['Themes'] + dx['descricao']+ dx['Título']
```

```python
dx.isnull().sum()
```

```python
df = dx.drop(columns=['Name_URL','Name', 'Name1', 'Score', 'Score2', "Genre", 'ReleaseDates1', 'GameModes', 'Themes', 'descricao', 'Título' ])
```

```python
df.columns = ['js', 'description']
df
```

```python
df['id'] = range(100)
df
```

```python
ds = df.drop(columns=['js'])
ds
```

```python
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(ds['description'].values.astype('U'))
```

```python
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
results = {}

for idx, row in ds.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities[idx][i], ds['id'][i]) for i in similar_indices]

    results[row['id']] = similar_items[1:]

print('done!')
```

```python
def item(id):
    return ds.loc[ds['id'] == id]['description'].tolist()[0].split(' - ')[0]
```

```python
##Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Recommending " + str(num) + " products similar to " + item(item_id) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: " + item(rec[1]) + " (score:" + str(rec[0]) + ")")
```

```python
recommend(item_id=3, num=5)
```
